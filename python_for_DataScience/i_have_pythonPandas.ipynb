#### df.columns = Veri setindeki tüm sütunların adlarını döndürür. 

```python
df.columns
```

## df.loc[satırNumarası,"sütunAdı"] = DataFrame'de sütunAdı ve satırNumarası ile belirtilen veriyi döndürürür. 

```python
df.loc[3,"fare"]
```

## df.loc[başlangıçDeğeri,bitişDeğeri] = başlangıç ve bitiş değeri ile arasındaki veri satırını seçmek için kullanabiliriz.

```python
df.loc[3:10]
```

## df.iloc[X] = X ile belirtilen satırdaki veriyi seçmek için kullanabiliriz. 

```python
df.iloc[1]
```

## df.iloc[[başlangıçDeğeri,bitişDeğeri]] = başlangıç ve bitiş değeri ile verilen veri satırını seçmek için kullanabiliriz.

```python
df.iloc[[3,10]] 
```

## df.iloc[başlangıçDeğeri:bitişDeğeri] = başlangıç değeri ile bitiş değeri(DAHİL DEĞİL!) arasındaki veri satırını seçmek için kullanabiliriz.

```python
df.iloc[3:10]
```

### DataFrame'de "x" isimli sütunundaki verilerin tipini değiştirmek için: df["x"] = df["x"].astype("veriTipi")

```python
df["survived"] = df["survived"].astype("float")
df
```

### describe() = İnteger tipindeki verilerin dağılımı hakkında istatistiksel (ortalama,standart Sapma,max değer,min değer...) gibi bilgiler almak için kullanılır

```python
df.describe().T
```

### rename() = Veri setinin sütun adlarını değiştirmek için kullanılır.

```python
df.rename(columns = {"survived" : "hayattaMi"
                    }, inplace = True)
df
```

### pop() = Veri setindeki belirli bir sütunu silmek için kullanılır.

```python
df.pop("pclass")
df
```

### drop() = Veri setinin belli bir satır/satırlar silmek için kullanılır

```python
df = df.drop([0],axis = 0)
df
```

### groupby() = Veri setini belli bir alana göre gruplandırmak ve istatistiksel analizler (ortalama,standart Sapma,max değer,min değer...) gibi yapmak için kullanılır.

```python
df.groupby(["class"]).mean()
```
